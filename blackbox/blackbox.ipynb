```python
from bs4 import BeautifulSoup
from lxml import html
import requests
import urllib3
import time

url = "http://95.85.18.95:8010/blackbox"
http_pool = urllib3.connection_from_url(url)
r = http_pool.urlopen('GET', url).data.decode()

responses = []

with requests.session() as s:
    resp = s.get(url)
    for i in range(-50, 51):
        try:
            post = s.post(url, data={'value':i})
            soup = BeautifulSoup(post.text, 'html.parser')
            field = soup.find('input', {'disabled':'disabled', 'type':'text'})
            responses.append((i, float(field['value'])))
            print(*responses[-1])
            time.sleep(1)
        except:
            print("Error occured at x = ", i)
```

$
\begin{align}
\text{Заметим, что в область определения искомой функции не входит только } x = 1 \\
\text{Тогда предположим, что искомой функцией будет являться }f(x) = \frac{k}{x + 1}\text{, где k - какое-то выражение}\\
\text{Найдем значение выражения }k\text{ при различных значениях }x\text{:}\\
x = -2 : k = 2\\
x = 9 : k = -9\\
x = 6: k = -6\\
\text{Заметим, что при любых значениях } x \text{значение искомого выражения будет равно } -x\\
\text{Значит, что искомая функция:}f(x) = \frac{-x}{x + 1}\\
\text{Проверим наше предположение}\\
\end{align}
$

```python
for response in responses:
        x, y = response
        yy = -x / (x + 1)
        print(x, y, yy, end = " ")
        if y == yy:
            print("OK")
        else:
            print("WRONG")
```
